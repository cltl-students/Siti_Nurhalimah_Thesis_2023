# OpenSubtitles
In this step, we:
1. extract data from OpenSubtitles corpus 

```python
file_paths = {
    'Arabic': './raw_data_opensubtitle/ar-id.dic',
    'Croatian': './raw_data_opensubtitle/hr-id.dic',
    'Thai': './raw_data_opensubtitle/id-th.dic',
    'Slovene': './raw_data_opensubtitle/id-sl.dic',
    'Finnish': './raw_data_opensubtitle/fi-id.dic',
    'Portuguese': './raw_data_opensubtitle/id-pt.dic',
    'Japanese': './raw_data_opensubtitle/id-ja.dic',
    'Polish': './raw_data_opensubtitle/id-pl.dic',
    'Greek': './raw_data_opensubtitle/el-id.dic',
    'Chinese': './raw_data_opensubtitle/id-zh.dic',
    'English': './raw_data_opensubtitle/en-id.dic',
    'Spanish': './raw_data_opensubtitle/es-id.dic'
}

translations = {}

for language, file_path in file_paths.items():
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            parts = line.strip().split('\t')
            if len(parts) == 6:
                if file_path.endswith('id.dic'):
                    translation = parts[2]
                    ind_word = parts[3]
                elif file_path.startswith('./raw_data_opensubtitle/id-'):
                    translation = parts[3]
                    ind_word = parts[2]
                elif file_path.startswith('./raw_data_bible/id-'):
                    translation = parts[3]
                    ind_word = parts[2]
                else:
                    translation = parts[2]
                    ind_word = parts[3]
                translations.setdefault(ind_word, {})[language] = translation
            if len(parts) < 6:
                if 'tanzil' in line:
                    translation = parts[1]
                    ind_word = parts[0]
                translations.setdefault(ind_word, {})[language] = translation

output_file = './data/open_subtitles.tsv'

with open(output_file, 'w', encoding='utf-8') as file:
    file.write('Indonesian\t' + '\t'.join(file_paths.keys()) + '\n')
    for ind_word, translation in translations.items():
        translation_values = [translation.get(language, 'None') for language in file_paths.keys()]
        file.write(f'{ind_word}\t' + '\t'.join(translation_values) + '\n')

print(f"Translations saved to {output_file}.")
```

# Merge all the corpus for parallel data OPUS
In this step, we:
1. merge all the corpus to build parallel data for OPUS

```python
def load_tsv_file(file_path):
    """
    Function to load data from a TSV (Tab-Separated Values) file.

    Param:
        file_path (str): The path to the TSV file to be loaded.
        
    Returns:
        dict: A dictionary containing the loaded data.
    """
    data = {}
    with open(file_path, 'r', encoding='utf-8') as tsv_file:
        reader = csv.reader(tsv_file, delimiter='\t')
        headers = next(reader)
        for row in reader:
            key = row[0]
            translations = row[1:]
            data[key] = translations
    return data

# Merge and update translations
def merge_translations(tsv1_data, tsv2_data, tsv3_data, tsv4_data):
    """
    Function to merge translations from multiple TSV datasets into a single dataset.
    
    Param:
        tsv1_data (dict): Translation data represented as a dictionary from open_subtitles corpus.
        tsv2_data (dict): Additional translation data from bible_uedin corpus to merge.
        tsv3_data (dict): Additional translation data from tanzil_qed corpus to merge.
        tsv4_data (dict): Additional translation data from qed corpus to merge.

    Returns:
        dict: A merged translation dataset with non-empty translations from all sources.
    """
    merged_data = tsv1_data.copy()
    for key, translations in tsv2_data.items():
        if key in merged_data:
            for i in range(len(translations)):
                if not merged_data[key][i]:
                    merged_data[key][i] = translations[i]
        else:
            merged_data[key] = translations

    for key, translations in tsv3_data.items():
        if key in merged_data:
            for i in range(len(translations)):
                if not merged_data[key][i]:
                    merged_data[key][i] = translations[i]
        else:
            merged_data[key] = translations

    for key, translations in tsv4_data.items():
        if key in merged_data:
            for i in range(len(translations)):
                if not merged_data[key][i]:
                    merged_data[key][i] = translations[i]
        else:
            merged_data[key] = translations

    return merged_data

# Load TSV files
tsv1_path = './data/open_subtitles.tsv'
tsv2_path = './data/bible_uedin.tsv'
tsv3_path = './data/tanzil_qed.tsv'
tsv4_path = './data/qed.tsv'

tsv1_data = load_tsv_file(tsv1_path)
tsv2_data = load_tsv_file(tsv2_path)
tsv3_data = load_tsv_file(tsv3_path)
tsv4_data = load_tsv_file(tsv4_path)

merged_data = merge_translations(tsv1_data, tsv2_data, tsv3_data, tsv4_data)

output_file = './data/output_opus.tsv'
with open(output_file, 'w', encoding='utf-8', newline='') as tsv_file:
    writer = csv.writer(tsv_file, delimiter='\t')
    writer.writerow(['Indonesian', 'Arabic', 'Croatian', 'Thai', 'Slovene', 'Finnish',
                     'Portuguese', 'Japanese', 'Polish', 'Greek', 'Chinese', 'English', 'Spanish'])
    for key, translations in merged_data.items():
        writer.writerow([key] + translations)

print("Translations merged and saved to", output_file)
```
